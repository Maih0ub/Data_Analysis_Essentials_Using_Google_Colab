## Day 1: Python Fundamentals and Visualization Basics

**Variables and Data Types**

```python
# Basic data types used in data analysis
company_revenue = 2500000       # Integer
growth_rate = 0.15              # Float
company_name = "TechCorp"       # String
is_profitable = True            # Boolean

print(f"Company: {company_name}")
print(f"Revenue: ${company_revenue:,}")
print(f"Growth Rate: {growth_rate:.1%}")
print(f"Profitable: {is_profitable}")
```

**Mathematical Operations**

```python
# Performing basic mathematical calculations
quantity = 100
price_per_unit = 25.50
total_revenue = quantity * price_per_unit
profit_margin = 0.20
profit = total_revenue * profit_margin

print(f"Quantity sold: {quantity}")
print(f"Price per unit: ${price_per_unit}")
print(f"Total revenue: ${total_revenue}")
print(f"Profit (20% margin): ${profit}")
```

**Lists**

```python
# Lists for storing related values
monthly_sales = [120000, 150000, 135000, 180000, 165000, 190000]
regions = ['North', 'South', 'East', 'West']
product_categories = ['Electronics', 'Fashion', 'Home', 'Beauty']

print("Monthly sales data:")
for i, sales in enumerate(monthly_sales):
    print(f"  Month {i+1}: ${sales:,}")

print(f"\nTotal 6-month sales: ${sum(monthly_sales):,}")
print(f"Average monthly sales: ${sum(monthly_sales) / len(monthly_sales):,.0f}")
print(f"Best month sales: ${max(monthly_sales):,}")
```

**Dictionaries**

```python
# Dictionaries store structured data with key-value pairs
customer_record = {
    'customer_id': 'CUST-001',
    'name': 'Sarah Johnson',
    'age': 32,
    'region': 'North',
    'total_spent': 2500.00,
    'premium_member': True,
    'purchase_count': 15
}

print("Customer Information:")
for key, value in customer_record.items():
    print(f"  {key}: {value}")

# Calculate average order value
avg_order_value = customer_record['total_spent'] / customer_record['purchase_count']
print(f"\nAverage order value: ${avg_order_value:.2f}")
```

**Conditionals**

```python
# Using if-elif-else to categorize customers
def categorize_customer(total_spent, purchase_count):
    if total_spent > 5000 and purchase_count > 20:
        return "VIP Customer"
    elif total_spent > 2000:
        return "Premium Customer"
    elif total_spent > 500:
        return "Regular Customer"
    else:
        return "New Customer"

# Test the function
test_customers = [
    {"name": "Alice", "spent": 7500, "purchases": 25},
    {"name": "Bob", "spent": 1200, "purchases": 8},
    {"name": "Carol", "spent": 300, "purchases": 2}
]

for customer in test_customers:
    category = categorize_customer(customer["spent"], customer["purchases"])
    print(f"{customer['name']}: {category} (${customer['spent']}, {customer['purchases']} purchases)")
```

**Loops**

```python
# Example of using loops to summarize quarterly performance
quarters = []
for i in range(0, len(monthly_sales), 3):
    quarter_sales = sum(monthly_sales[i:i+3])
    quarters.append(quarter_sales)
    print(f"Q{len(quarters)} Sales: ${quarter_sales:,}")

print(f"Best performing quarter: Q{quarters.index(max(quarters)) + 1}")
```

**Functions**

```python
# Function to calculate key sales metrics
def calculate_metrics(sales_data):
    """Return total, average, growth, best and worst sales values"""
    total = sum(sales_data)
    average = total / len(sales_data)
    growth = ((sales_data[-1] - sales_data[0]) / sales_data[0]) * 100

    return {
        'total': total,
        'average': average,
        'growth_rate': growth,
        'best_month': max(sales_data),
        'worst_month': min(sales_data)
    }

# Apply the function
metrics = calculate_metrics(monthly_sales)

print("Sales Performance Metrics:")
print(f"  Total Sales: ${metrics['total']:,}")
print(f"  Average Monthly: ${metrics['average']:,.0f}")
print(f"  Growth Rate: {metrics['growth_rate']:.1f}%")
print(f"  Best Month: ${metrics['best_month']:,}")
print(f"  Worst Month: ${metrics['worst_month']:,}")
```

**Importing Pandas Library**

```python
# Import pandas library for data manipulation
import pandas as pd
import warnings

# Suppress warnings for cleaner output
warnings.filterwarnings('ignore')
```

**Creating a Sample Dataset**

```python
# Create a Python dictionary to hold monthly sales data
sales_data = {
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun'],
    'Sales': [120000, 150000, 135000, 180000, 165000, 190000],
    'Region': ['North', 'South', 'East', 'West', 'North', 'South'],
    'Target': [125000, 140000, 130000, 175000, 160000, 185000]
}

# Convert the dictionary to a pandas DataFrame
df = pd.DataFrame(sales_data)
```

**Exploring the DataFrame**

```python
# Basic DataFrame information
print("DataFrame Information:")
print(f"Shape: {df.shape} (rows, columns)")
print(f"Column names: {list(df.columns)}")

# Display the first few rows of the DataFrame
print("\nFirst 5 rows of the DataFrame:")
display(df.head())
```

**Adding New Calculated Columns**

```python
# Add columns for target achievement and variance
df['Target_Achievement'] = (df['Sales'] / df['Target'] * 100).round(1)
df['Variance'] = df['Sales'] - df['Target']

print("Enhanced DataFrame with calculations:")
df
```

**Summary Statistics**

```python
# Basic statistical summaries
print("Summary Statistics:")
print(f"Average Sales: ${df['Sales'].mean():,.0f}")
print(f"Total Sales: ${df['Sales'].sum():,}")
print(f"Best Month: {df.loc[df['Sales'].idxmax(), 'Month']} (${df['Sales'].max():,})")
```

#Fast Mathematical Operations

**Importing NumPy**

```python
# Import NumPy for numerical operations
import numpy as np
```

**Converting Lists to NumPy Arrays**

```python
# Convert sales data to NumPy arrays
sales_array = np.array(df['Sales'])
target_array = np.array(df['Target'])

print("Sales data as NumPy array:", sales_array)
print("Array shape:", sales_array.shape)
print("Array data type:", sales_array.dtype)
```

**NumPy Statistical Operations**

```python
# Perform basic statistical calculations
print("NumPy Statistical Summary:")
print(f"Mean: ${np.mean(sales_array):,.2f}")
print(f"Median: ${np.median(sales_array):,.2f}")
print(f"Standard Deviation: ${np.std(sales_array):,.2f}")
print(f"Min: ${np.min(sales_array):,}")
print(f"Max: ${np.max(sales_array):,}")
```

**Element-Wise Operations**

```python
# Perform element-wise operations
variance_array = sales_array - target_array
achievement_rate = (sales_array / target_array) * 100

print("Variance from target:", variance_array)
print("Achievement rates (%):", achievement_rate.round(1))
```

**Growth Rate Calculation**

```python
# Calculate month-over-month growth
growth_rates = np.diff(sales_array) / sales_array[:-1] * 100
print("Month-over-month growth rates (%):", growth_rates.round(1))
```

**Conditional Operations**

```python
# Create a boolean array to check if sales exceeded targets
over_target = sales_array > target_array
print("Months that beat target:", np.sum(over_target), "out of", len(sales_array))
print(f"Percentage over target: {(np.sum(over_target) / len(over_target) * 100):.1f}%")
```

#Matplotlib - Visualizations

**Import Matplotlib**

```python
# Import Matplotlib for visualizations
import matplotlib.pyplot as plt
```

**Line Chart: Monthly Sales vs Target**

```python
# Create a line chart comparing actual sales vs target
plt.figure(figsize=(8, 6))
plt.plot(df['Month'], df['Sales'], marker='o', linewidth=2, color='#2E86AB', label='Actual Sales')
plt.plot(df['Month'], df['Target'], marker='s', linewidth=2, linestyle='--', color='#A23B72', label='Target')
plt.title('Monthly Sales vs Target', fontweight='bold')
plt.xlabel('Month')
plt.ylabel('Sales ($)')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

**Bar Chart: Target Achievement by Month**

```python
# Create a bar chart with conditional colors based on achievement
plt.figure(figsize=(8, 6))
colors = ['green' if x >= 100 else 'red' for x in df['Target_Achievement']]
plt.bar(df['Month'], df['Target_Achievement'], color=colors, alpha=0.7)
plt.axhline(y=100, color='black', linestyle='--', linewidth=2, label='Target (100%)')
plt.title('Target Achievement by Month', fontweight='bold')
plt.xlabel('Month')
plt.ylabel('Achievement (%)')
plt.legend()
plt.grid(True, alpha=0.3, axis='y')
plt.show()
```

**Regional Sales Comparison**

```python
# Group by region and visualize total sales
region_sales = df.groupby('Region')['Sales'].sum().sort_values(ascending=False)

plt.figure(figsize=(8, 6))
plt.bar(region_sales.index, region_sales.values, color=['#FF6B35', '#F7931E', '#FDC830', '#37B5A6'])
plt.title('Total Sales by Region', fontweight='bold')
plt.xlabel('Region')
plt.ylabel('Total Sales ($)')

# Add value labels on top of each bar
for i, v in enumerate(region_sales.values):
    plt.text(i, v + 5000, f'${v:,.0f}', ha='center', fontweight='bold')

plt.show()
```

**Pie Chart: Regional Sales Distribution**

```python
# Create a pie chart showing sales distribution by region
plt.figure(figsize=(8, 6))
plt.pie(region_sales.values, labels=region_sales.index, autopct='%1.1f%%',
        colors=['#FF6B35', '#F7931E', '#FDC830', '#37B5A6'], startangle=90)
plt.title('Regional Sales Distribution', fontweight='bold')
plt.show()
```

**Checking Data Quality**

```python
# Check for missing values
print("--- Missing Values ---")
missing = df.isnull().sum()
print(missing[missing > 0])

# Check for duplicates
duplicates = df.duplicated().sum()
print(f"\nDuplicate rows: {duplicates}")
```

**Handling Missing Values**

```python
# Fill missing ages with the median
df['Age'].fillna(df['Age'].median(), inplace=True)
print("Missing ages have been filled using the median value.")
print(f"Remaining missing values in Age: {df['Age'].isnull().sum()}")
```

**Summary Statistics**

```python
# Summary statistics for numerical columns
print("--- Summary Statistics ---")
print(df.describe().round(2))

# Key metrics
print(f"\nAverage Price: ${df['Price'].mean():.2f}")
print(f"Total Revenue: ${df['TotalAmount'].sum():,.2f}")
print(f"Average Customer Age: {df['Age'].mean():.1f} years")
```

**Customer & Transaction Overview**

```python
# Categorical distributions
print("--- Gender Distribution ---")
print(df['Gender'].value_counts())

print("\n--- Region Distribution ---")
print(df['Region'].value_counts())

print("\n--- Product Category Distribution ---")
print(df['Category'].value_counts())

print("\n--- Payment Method Distribution ---")
print(df['PaymentMethod'].value_counts())
```

**Most Expensive Product in Each Category**

```python
most_expensive = df.loc[df.groupby('Category')['Price'].idxmax(), ['Category', 'Product', 'Price']]
most_expensive = most_expensive.sort_values('Category').reset_index(drop=True)
print("Most Expensive Product in Each Category:")
print(most_expensive)
```

**Average Age by Region**

```python
average_age_region = df.groupby('Region')['Age'].mean().round(1).reset_index()
print("Average Age by Region:")
print(average_age_region)
```

**Most Popular Product Category per Region**

```python
category_counts = df.groupby(['Region', 'Category']).size().reset_index(name='Count')
most_popular_by_region = category_counts.loc[category_counts.groupby('Region')['Count'].idxmax()]
most_popular_by_region = most_popular_by_region.sort_values('Region').reset_index(drop=True)

print("Most Popular Product Category in Each Region:")
print(most_popular_by_region)
```

**Visualization: Age Distribution**

```python
# Plot age distribution
plt.figure(figsize=(10, 6))
plt.hist(df['Age'], bins=25, color='skyblue', edgecolor='black')
plt.axvline(df['Age'].mean(), color='red', linestyle='--', linewidth=2, label=f'Average: {df["Age"].mean():.1f}')
plt.xlabel('Age (years)')
plt.ylabel('Number of Customers')
plt.title('Customer Age Distribution')
plt.legend()
plt.grid(axis='y', alpha=0.3)
plt.show()
```

**Visualization: Sales by Category**

**Compare Average Transaction Value by Gender**

```python
avg_transaction_gender = df.groupby('Gender')['TotalAmount'].mean().round(2).reset_index()
print("Average Transaction Value by Gender:")
print(avg_transaction_gender)

plt.figure(figsize=(8, 6))
sns.barplot(data=avg_transaction_gender, x='Gender', y='TotalAmount', palette='pastel')
plt.title('Average Transaction Value by Gender')
plt.ylabel('Average Transaction Value ($)')
plt.show()
```

## Day 3: Regional Revenue Analysis and Summary Report

**Calculating Revenue by Region**

**Visualizing Revenue Distribution by Region**

```python
# Bar chart of revenue by region
plt.figure(figsize=(10, 6))
sns.barplot(data=region_revenue, x='Region', y='TotalAmount', palette='coolwarm')
plt.title('Total Revenue by Region', fontweight='bold')
plt.xlabel('Region')
plt.ylabel('Total Revenue ($)')

# Add value labels on each bar
for i, v in enumerate(region_revenue['TotalAmount']):
    plt.text(i, v + (0.01 * v), f'${v:,.0f}', ha='center', fontweight='bold')

plt.tight_layout()
plt.show()
```

**Pie Chart: Revenue Percentage by Region**

```python
# Pie chart showing revenue contribution by region
plt.figure(figsize=(8, 8))
plt.pie(region_revenue['TotalAmount'], labels=region_revenue['Region'], autopct='%1.1f%%',
        startangle=90, colors=sns.color_palette('husl'))
plt.title('Revenue Share by Region', fontweight='bold')
plt.tight_layout()
plt.show()
```
